```python
import subprocess, pathlib, glob, tqdm

shots = ['sweep','hook','pull','flick','cover','defense',
         'lofted','straight','square_cut','late_cut']

SRC_ROOT = pathlib.Path('/dataset')
DST_ROOT = pathlib.Path('dataset_256p')

for shot in shots:
    in_dir  = SRC_ROOT / shot
    out_dir = DST_ROOT / shot
    out_dir.mkdir(parents=True, exist_ok=True)

    for vid_path in tqdm.tqdm(in_dir.glob('*.avi'), desc=shot):
        out_path = out_dir / f'{vid_path.stem}_256p.mp4'
        if out_path.exists():          # skip already‑converted files
            continue

        subprocess.run([
            'ffmpeg', '-i', str(vid_path),
            '-vf', 'fps=25,scale=-2:256,setsar=1',
            '-c:v', 'libx264', '-crf', '22', '-preset', 'fast',
            '-an', str(out_path)
        ], check=True)
print("✅  Finished every class.")
```
